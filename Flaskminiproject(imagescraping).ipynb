```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen as uReq
import os 
```

```python
save_dir="images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
query="Virat kohli"
response=requests.get(f"https://www.google.com/search?q={query}sudhansu+kumar+ineuron&tbm=isch&ved=2ahUKEwiGkNf6rdqDAxXUl2MGHefyDYUQ2-cCegQIABAA&oq=sudhansu+kumar+ineuron&gs_lcp=CgNpbWcQAzoECCMQJzoHCAAQgAQQGDoECAAQHlDyA1jQFmDFG2gAcAB4AIABmQGIAdYJkgEDMC45mAEAoAEBqgELZ3dzLXdpei1pbWfAAQE&sclient=img&ei=KYGiZcbXB9SvjuMP5-W3qAg&bih=703&biw=1536&hl=en")
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
images_tages=soup.find_all("img")
```

```python
for i in images_tages:
    images_url=i['src']
    image_data=requests.get(images_url).content
    with open(os.path.join(save_dir, f"{query}_{images_tages.index(i)}.jpg"),"wb") as f:
        f.write(image_data)
    
```
